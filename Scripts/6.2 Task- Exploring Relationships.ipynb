# 05. Revisiting Project Questions and Creating Hypotheses

### What is the most common cause of accidents?
>#### The most common cause of accidents for the primary vehicle is that the driver is driving unsafe/inexperienced/at fault.

### What type of cars are most commonly in accidents?
>#### Sedan/SUV's are the car that are most commonly in accidents

### What contributing factors lead to the most accident injuries? Deaths?
>#### The most accident injuries occur when the primary vehicle's contributing factors are Drug/Alcohol Consumption, the driver is unsafe/inexperienced/at fault, or there were external distractions/causes.
>#### The most accident deaths occur whent the primary vehicle's contributing factors are external distractions.

### Potential Hypothesis
>#### When a driver consumes drugs or alcohol, their likelihood of being in an accident where a person is injured or dies increases.
>#### There is an uptick in dangerous accidents during holiday seasons because people are in a rush to get where they need to be, and drive more distracted.
>#### Accident rates increase more in the mornings during traffic hour when people are on their ways to work.
